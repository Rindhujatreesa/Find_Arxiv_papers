```python
# import all the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import re
```

### Plotting Pie charts on the percentage of number of authors per paper

```python
#The labels required for the pie chart are created as a list
labels = ['Single author','2 authors','3 - 4 authors','More than 4 authors']
```

```python
#the csv file is converted into a pandad DataFrame using pd.read_csv
databases = pd.read_csv('Databases.csv',index_col = "Serial No.")
authors = list(databases["Authors"]) #the Authors column is converted into list to enable the counting of number of authors 
no_of_authors_DB = []
for author in authors:
    count = 0
    names = re.findall("(\w+ (\w+. )?\w+)+",author) #we get a list of author names for each paper
    count = len(names) #the length of the names list will give the number of authors of the paper
    no_of_authors_DB.append(count) #no_of_authors_DB list will have a list of numbers which denotes no, of authors of each paper
databases['Number_of_authors'] = no_of_authors_DB #we add the number of authors as a new column to the df
databases.head()
```

```python
#We initialize four variables which will store the number of papers with on author, 2 authors, 3 to 4 authors,
# and more than 4 authors
single_author = 0
two_authors = 0
three_four_authors = 0
more_than_four = 0
#We loop through the list that has the number of authors for each paper and uses the if-elif statements 
# to add count to each variable
for no_of_authors in no_of_authors_DB:
    if no_of_authors == 1:
        single_author += 1
        #print(single_author)
    elif no_of_authors == 2:
        two_authors +=1
        #print(two_authors)
    elif (no_of_authors >2) & (no_of_authors < 5):
        three_four_authors += 1
        #print(three_four_authors)
    else:
        more_than_four +=1
        #print(more_than_four)
Authors_DB = [single_author,two_authors,three_four_authors, more_than_four] #a list is created with the counts of each label
```

```python
Authors_DB
```

```python
#we plot the pie chart using the matplotlib.pyplot
fig1, ax1 = plt.subplots()
ax1.pie(Authors_DB, labels=labels, autopct='%1.2f%%',
        shadow=True, startangle=0)
ax1.axis('equal')
ax1.set_title('Percentage of number of Authors in Database Category')
plt.show()
```
