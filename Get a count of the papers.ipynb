```python
# import all the necessary libraries
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
```

#### Bar Graph of number of papers published in each category in each year
##### For each of the primary categories, we create a list which contains the number of papers published in the years 2018,2019,2020,2021, and 2022

NB: There are a few papers which were added in the arxiv in 2018, however, was prepared in 2017. Since, in this project the metadata extraction was done using the arxiv ID of the format YYMM.XXXXX, all the papers with ids starting with 18,19,20,21,22 are included.

```python
#extracting data from csv file to a dataframe
databases = pd.read_csv('Databases.csv',index_col = 'Serial No.')
#We use the published dates to extract the year from each paper
published_date = list(databases['Published On'])
years = []
for date in published_date:
    date_ = datetime.fromisoformat(date[:-6]) #we extract the year portion from the date using fromisoformat() function
    years.append(date_.year)
databases['Year'] = years
#databases.head()
# year variables are defined which would store the number of papers added to the arxiv in each year
_2018 = 0
_2019 = 0
_2020 = 0
_2021 = 0
_2022 = 0
# We loop the years list and adds to the appropriate variable when a match for the year is found
for year in years:
    if year == 2018:
        _2018 +=1
    elif year == 2019:
        _2019 +=1
    elif year == 2020:
        _2020 +=1
    elif year == 2021:
        _2021 +=1
    elif year == 2022:
        _2022 +=1
No_of_books_DB = [_2018,_2019,_2020,_2021,_2022] #This list will be used for plotting the bar graphs
No_of_books_DB
```

```python
#extracting data from csv file to a dataframe
graphics = pd.read_csv('Graphics.csv',index_col = 'Serial No.')
#We use the published dates to extract the year from each paper
published_date = list(graphics['Published On'])
years = []
for date in published_date:
    date_ = datetime.fromisoformat(date[:-6])
    years.append(date_.year)
graphics['Year'] = years
#graphics.tail()
# year variables are defined which would store the number of papers added to the arxiv in each year
_2018 = 0
_2019 = 0
_2020 = 0
_2021 = 0
_2022 = 0
# We loop the years list and adds to the appropriate variable when a match for the year is found
for year in years:
    if year == 2018:
        _2018 +=1
    elif year == 2019:
        _2019 +=1
    elif year == 2020:
        _2020 +=1
    elif year == 2021:
        _2021 +=1
    elif year == 2022:
        _2022 +=1
No_of_books_GR = [_2018,_2019,_2020,_2021,_2022] #This list will be used for plotting the bar graphs
No_of_books_GR
```

```python
#extracting data from csv file to a dataframe
robotics = pd.read_csv('Robotics.csv',index_col = 'Serial No.')
#We use the published dates to extract the year from each paper
published_date = list(robotics['Published On'])
years = []
for date in published_date:
    date_ = datetime.fromisoformat(date[:-6])
    year = date_.year
    if year == 2017: #this makes sure that we do not miss out the papers that were published in 2017 but added to arxiv in 2018
        year = 2018
    years.append(year)
robotics['Year'] = years
#robotics.tail()
# year variables are defined which would store the number of papers added to the arxiv in each year
_2018 = 0
_2019 = 0
_2020 = 0
_2021 = 0
_2022 = 0
# We loop the years list and adds to the appropriate variable when a match for the year is found
for year in years:
    if year == 2018:
        _2018 +=1
    elif year == 2019:
        _2019 +=1
    elif year == 2020:
        _2020 +=1
    elif year == 2021:
        _2021 +=1
    elif year == 2022:
        _2022 +=1
No_of_books_RO = [_2018,_2019,_2020,_2021,_2022] #This list will be used for plotting the bar graphs
No_of_books_RO
```

```python
#extracting data from csv file to a dataframe
ET = pd.read_csv('Emerging Technologies.csv',index_col = 'Serial No.')
#We use the published dates to extract the year from each paper
published_date = list(ET['Published On'])
years = []
for date in published_date:
    date_ = datetime.fromisoformat(date[:-6])
    year = date_.year
    if year == 2017:#this makes sure that we do not miss out the papers that were published in 2017 but added to arxiv in 2018
        year = 2018
    years.append(year)
ET['Year'] = years
#ET.tail()
# year variables are defined which would store the number of papers added to the arxiv in each year
_2018 = 0
_2019 = 0
_2020 = 0
_2021 = 0
_2022 = 0
# We loop the years list and adds to the appropriate variable when a match for the year is found
for year in years:
    if year == 2018:
        _2018 +=1
    elif year == 2019:
        _2019 +=1
    elif year == 2020:
        _2020 +=1
    elif year == 2021:
        _2021 +=1
    elif year == 2022:
        _2022 +=1
No_of_books_ET = [_2018,_2019,_2020,_2021,_2022] #This list will be used for plotting the bar graphs
No_of_books_ET
```

```python
Years = [2018,2019,2020,2021,2022] #the x-labels for the bar graph
fig = plt.figure(figsize = (40,10)) #a plot is defined with the specific dimensions
ax = plt.axes()
x = np.arange(len(Years)) #we define a array of numbers i.e.,(0,1,2,3,4)
width = 0.2 #the width of each bar is defined
#We have four paper categories for each year. We allocate equal bar width for each of them for all the years
# (x - 2*width),(x - width),x, and (x + width) will be the four points where each bar begins and will have a width equals to 0.2
ax.bar(x - 2*width, No_of_books_DB, width, color="#36355c",label="Databases")
ax.bar(x - width, No_of_books_GR, width, color="#b64e7e",label="Graphics")
ax.bar(x,No_of_books_RO,width, color = '#ff8d52', label = "Robotics")
ax.bar(x + width,No_of_books_ET,width, color = '#f4ff2f', label = "Emerging Tech")
ax.set_xticks(x) #The xticks will be the array defined before
ax.set_xticklabels(Years)#the ticks will be labeled as years
ax.set_xlabel("YEAR",fontsize = 24)
ax.set_ylabel("Number of Papers",fontsize = 24)
ax.set_title("Number of Papers in arxiv from 2018 to 2022",fontsize = 32)
plt.tick_params(axis = 'x', rotation = 45,labelsize  = 24)
plt.tick_params(axis = 'y', rotation = 0,labelsize  = 24)
plt.legend(fontsize = 24) #the legend appears on the graph
plt.show()
```
