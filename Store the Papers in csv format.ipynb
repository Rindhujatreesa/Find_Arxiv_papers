```python
# import all the necessary libraries
import pandas as pd
import re
```

### Retrieving Data from csv file into a Pandas DataFrame (Databases.csv)

```python
# the csv file is converted to a dataframe using pd.read_csv()
df_DB = pd.read_csv('Databases.csv',index_col = "Serial No.")
# The authors column in the csv file is of the format arxiv.Search.author('author_name')
# Using the below for loop and regex pattern, the names are filtered out and it edited in the csv file as well
Authors = list(df_DB['Authors']) 
authors = []
for author in Authors:
    names = re.findall("(\w+ (\w+. )?\w+)+",author) #Gets the name found in the parenthesis
    authors.append(names)
df_DB['Authors'] = authors
df_DB.to_csv('Databases.csv',index = True, index_label = "Serial No.") #The already existing Authors column is modified
df_DB.head()
```
